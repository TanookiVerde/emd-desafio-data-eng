```python
import requests
import pandas as pd
```

## Recuperação de Dados

```python
response = requests.get(
    "https://dados.mobilidade.rio/gps/brt"
)

content = response.json()

dados = pd.DataFrame(content['veiculos'])
dados.sort_values(by='codigo',inplace=True)

dados.head()
```

## Limpeza Inicial

### Tratamento para Identificação de Valores Nulos

```python
dados_tratados = dados.replace("", pd.NA).replace(" ", pd.NA)

dados_tratados.info()
```

```python
dados_tratados.describe()
```

## Análise de Distribuição

### Sentido

```python
dados['quantidade'] = '-'
sentido_distribuicao = dados.groupby(by='sentido').count()[['quantidade']]

sentido_distribuicao
```
